###2- Evaluation du modèle d'apprentissage sur les

```python
import numpy as np  #Importation bibliothèque numpy
from sklearn.linear_model import LinearRegression   #Importation du "LinearRegression"
regression = LinearRegression()
regression.fit(features, valeurs_cibles) #Entraînement du modèle - On obtient les labels Réels
valeurs_cibles_predites=regression.predict(features)#Obtentions des valeurs prédites
MSE=sk.metrics.mean_squared_error(valeurs_cibles, valeurs_cibles_predites) #Calcul du MSE pour le modèle linéaire
RMSE=np.sqrt(MSE)   #Calcul du RMSE pour le modèle linéaire
print(RMSE)
```

```python
from sklearn.tree import DecisionTreeRegressor #Importation du modèle "DecisionTreeRegressor"
tree=sk.tree.DecisionTreeRegressor()
tree.fit(features, valeurs_cibles)  #Entraînement du modèle sur les valeurs
valeurs_cibles_predites_tree=tree.predict(features, check_input=True)
MSE=sk.metrics.mean_squared_error(valeurs_cibles, valeurs_cibles_predites_tree) #Calcul du MSE pour le modèle linéaire
RMSE=np.sqrt(MSE) #Calcul du RMSE pour le modèle linéaire
print(RMSE)# On obtient une RMSE en 0 cela veut dire que pour chaque observation, l'arbre de décision prédit exactement la bonne valeur.
```

#2- Evaluation du modèle d'apprentissage sur les données de test

37

```python
#On refait exactement pareil mais cette fois pour les données de test
import pandas as pd
import numpy as np
import sklearn as sk

# Charger les données de test
featurestest = test_christian.drop(columns='median_house_value', axis=1)
valeurs_ciblestest = test_christian['median_house_value'].copy()

# Remplir les valeurs manquantes pour 'total_bedrooms' par la médiane
calcul_median_test = featurestest['total_bedrooms'].median()
featurestest['total_bedrooms'].fillna(calcul_median_test)

# Encodage de la variable 'ocean_proximity'
encodeur = sk.preprocessing.OrdinalEncoder()
proxi_ocean_test = featurestest[["ocean_proximity"]]
christian_encode_test = encodeur.fit_transform(proxi_ocean_test)
featurestest["ocean_proximity"] = christian_encode_test


# Récupérer le meilleur modèle après la recherche par grille
param_gridtest = [{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]}]
random_forest_regressortest = RandomForestRegressor()
grid_searchtest = GridSearchCV(estimator=random_forest_regressortest, param_grid=param_gridtest, scoring='neg_mean_squared_error', cv=3)
grid_searchtest.fit(featurestest, valeurs_ciblestest)
best_model = grid_searchtest.best_estimator_
print("Le modèle le plus adapté est :")
print(best_model)

# Prédictions sur la base de test
valeurs_cibles_predites_test = best_model.predict(featurestest)
print(valeurs_cibles_predites_test)

# Calcul du MSE et du RMSE
MSE_test = sk.metrics.mean_squared_error(valeurs_ciblestest, valeurs_cibles_predites_test)
RMSE_test = np.sqrt(MSE_test)
print("RMSE sur la base de test :")
print(RMSE_test)






```

Ce RMSE calculé à l'aide du meilleur modèle de RansomForestRegressor est d'environ 21000 nous permet de conclure que celui-ci est assez performant car les valeurs cibles prédictes varient dans une plage plus large(Les valeurs de MedianHouseValue sont en moyenne de 210000 (Q10)) . Donc on peut dire qu'un RMSE d'une valeur d'environ 21000 est un bon résultat car elle correspond a une erreur d'environ 10%.
